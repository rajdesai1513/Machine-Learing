```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("Iris.csv")
df
```

```python
sns.heatmap(df.iloc[:,:-1].corr())
plt.savefig("corelation.png")
```

```python
plt.grid()
plt.xlabel("SepalLenghtCm")
plt.ylabel("SepalWidthCm")
plt.plot(df["SepalLengthCm"],df["SepalWidthCm"])
plt.savefig("linechart.png")
```

```python
sns.histplot(data=df,x="Species")
plt.savefig("histplot.png")
```

```python
plt.title("BoxPlot")
plt.savefig("sepal.jpg")
sns.boxplot(x=df["SepalLengthCm"])
plt.savefig("Heloo.png")
```

```python
plt.scatter(df["SepalLengthCm"],df["Species"])
plt.savefig("scatter.png")
```
